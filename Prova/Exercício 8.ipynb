```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

```python
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```python
X = tf.placeholder(tf.float32, shape=[4,2], name = 'X')
Y = tf.placeholder(tf.float32, shape=[4,1], name = 'Y')

W = tf.Variable(tf.truncated_normal([2,2]), name = "W")
w = tf.Variable(tf.truncated_normal([2,1]), name = "w")

c = tf.Variable(tf.zeros([4,2]), name = "c")
b = tf.Variable(tf.zeros([4,1]), name = "b")
```

# Gradient Descendent Optimizer

```python
INPUT_XOR = [[0,0],[0,1],[1,0],[1,1]]
OUTPUT_XOR = [[0],[1],[1],[0]]

init = tf.global_variables_initializer()
sess = tf.Session()

writer = tf.summary.FileWriter("./logs/xor_logs", sess.graph)

sess.run(init)
```

```python
with tf.name_scope("hidden_layer") as scope:
    h = tf.nn.relu(tf.add(tf.matmul(X, W),c))

with tf.name_scope("output") as scope:
    y_estimated = tf.sigmoid(tf.add(tf.matmul(h,w),b))

with tf.name_scope("erro_GD") as scope:
    erro_GD = tf.reduce_mean(tf.squared_difference(y_estimated, Y)) 

with tf.name_scope("train") as scope:
    train_step = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(erro_GD)
```

```python
for epoch in range(40001):
    sess.run(train_step, feed_dict={X: INPUT_XOR, Y: OUTPUT_XOR})
    if epoch % 100 == 0:
        print("_"*80)
        print('Epoch: ', epoch)
        print('   erro_GD: ', sess.run(erro_GD, feed_dict={X: INPUT_XOR, Y: OUTPUT_XOR}))
```

```python
reset_graph()
```

# Adam Optimizer

```python
X = tf.placeholder(tf.float32, shape=[4,2], name = 'X')
Y = tf.placeholder(tf.float32, shape=[4,1], name = 'Y')

W = tf.Variable(tf.truncated_normal([2,2]), name = "W")
w = tf.Variable(tf.truncated_normal([2,1]), name = "w")

c = tf.Variable(tf.zeros([4,2]), name = "c")
b = tf.Variable(tf.zeros([4,1]), name = "b")
```

```python
with tf.name_scope("hidden_layer") as scope:
    h = tf.nn.relu(tf.add(tf.matmul(X, W),c))

with tf.name_scope("output") as scope:
    y_estimated = tf.sigmoid(tf.add(tf.matmul(h,w),b))

with tf.name_scope("erro_AdOpt") as scope:
    erro_AdOpt = tf.reduce_mean(tf.squared_difference(y_estimated, Y)) 

with tf.name_scope("train") as scope:
    train_step = tf.train.AdamOptimizer(learning_rate=0.01).minimize(erro_AdOpt)
```

```python
INPUT_XOR2 = [[0,0],[0,1],[1,0],[1,1]]
OUTPUT_XOR2 = [[0],[1],[1],[0]]

init = tf.global_variables_initializer()
sess = tf.Session()

writer = tf.summary.FileWriter("./logs/xor_logs", sess.graph)

sess.run(init)
```

```python
for epoch in range(40001):
    sess.run(train_step, feed_dict={X: INPUT_XOR, Y: OUTPUT_XOR})
    if epoch % 100 == 0:
        print("_"*80)
        print('Epoch: ', epoch)
        print('   erro_AdOpt: ', sess.run(erro_AdOpt, feed_dict={X: INPUT_XOR, Y: OUTPUT_XOR}))
```
